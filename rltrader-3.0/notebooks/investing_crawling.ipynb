# investing.com 크롤링으로 차트 데이터 획득하기

https://kr.investing.com/equities/samsung-electronics-co-ltd-historical-data

```python
from io import StringIO
import requests
import pandas as pd
url = 'https://kr.investing.com/instruments/HistoricalDataAjax'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
        'AppleWebKit/537.36 (KHTML, like Gecko) '
        'Chrome/96.0.4664.110 Whale/3.12.129.46 Safari/537.36',
    'x-requested-with': 'XMLHttpRequest',
}
data = {
    'curr_id': '43433',
    'smlID': '25193974',
    'header': '005930 역사적 데이터',
    'st_date': '2021/12/01',
    'end_date': '2021/12/31',
    'interval_sec': 'Daily',
    'sort_col': 'date',
    'sort_ord': 'DESC',
    'action': 'historical_data',
}
res = requests.post(url, data, headers=headers)

list_df = pd.read_html(StringIO(res.text))
df = list_df[0]
df.head()
```
